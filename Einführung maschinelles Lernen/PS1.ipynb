```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import datasets
X, y= datasets.make_blobs(centers=2,n_samples=20)
y = [val-1 if val == 0 else val for val in y]
print(y)
```

```python
def perceptron(X,y,Learn):
    w=np.array([0,0])
    b = 0
    R = max([np.linalg.norm(val) for val in X])
    counter = 0
    made_mistake = True
    while made_mistake:
        made_mistake = False
        for num,val in enumerate(X):
             if y[num] * (np.dot(val,w)+b) <= 0:
                x1 = np.linspace(-10,10,300)
                x2 = []
                w = w + (Learn * y[num] * val)
                b = b + (Learn * y[num] * (R**2))
                counter += 1
                made_mistake = True
                for x in x1:
                    x2.append((-w[0]*x)/w[1] - b / w[1])
                plt.scatter(x1,x2,marker = "_",lw=1,color=plt.cm.cool(counter)) 
    return (w,b,counter)
```

```python
from sklearn import datasets
X, y= datasets.make_blobs(centers=2,n_samples=20)
y = [val-1 if val == 0 else val for val in y]
print(y)
plt.figure(figsize=(12,5))
w, b, counter = perceptron(X,y,0.1)
print(w)
print(b)
print(counter)

X1 = [X[i][0] for i in range(len(X))]
X2 = [X[i][1] for i in range(len(X))]

plt.scatter(X1,X2,c=y,)
```
