```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X1 = np.linspace(-1,1,50)
X2 = np.linspace(-1,1,50)
```

```python
def points(X1,X2):
    points = []
    for x in X1:
        for y in X2:
            if x**2+y**2<0.5 and x**2+y**2>-0.5:
                color ="red"
                points.append([x,y,color])
            else:
                color ="blue"
                points.append([x,y,color])
    return points

def plot(points):
    for point in points:
        plt.scatter(point[0],point[1],c=point[2],s=4)

def transform_xx(points):
    trans_point = []
    for point in points:
        trans_x = point[0]**2
        trans_y = point[1]**2
        trans_point.append([trans_x,trans_y,point[2]])
    return trans_point
```

```python
Grid=points(X1,X2)
plot(Grid)
```

```python
Grid_trans = transform_xx(Grid)
plot(Grid_trans)
```
