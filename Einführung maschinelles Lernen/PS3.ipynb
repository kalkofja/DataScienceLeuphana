```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Task I
X1 = np.linspace(-1,1,50)
X2 = np.linspace(-1,1,50)
```

```python
def points(X1,X2):
    points = []
    for x in X1:
        for y in X2:
            if x**2+y**2<0.5 and x**2+y**2>-0.5:
                color ="red"
                points.append([x,y,color])
            else:
                color ="blue"
                points.append([x,y,color])
    return points

def plot(points):
    for point in points:
        plt.scatter(point[0],point[1],c=point[2],s=4)

def transform_xx(points):
    trans_point = []
    for point in points:
        trans_x = point[0]**2
        trans_y = point[1]**2
        trans_point.append([trans_x,trans_y,point[2]])
    return trans_point
```

```python
Grid=points(X1,X2)
plt.figure(figsize=(5,5))
plot(Grid)
```

```python
Grid_trans = transform_xx(Grid)
plt.figure(figsize=(5,5))
plot(Grid_trans)
```

```python
#Task II
A = [-1,-1,1,1]
B = [-1,1,-1,1]
```

```python
def boolean_AorB (A,B):
    for i in range(len(A)):
        x1 = A[i]
        x2 = B[i] 
        x = np.array([1,x1,x2])
        w1=np.array([1.5,1,1]) 
        if np.dot(x,w1) < 0.5:
            print("A= " + str(A[i]) + " or " + "B= " + str(B[i]) + " = False")
        else:
            print("A= " + str(A[i]) + " or " + "B= " + str(B[i]) + " = True")

def boolean_AandB (A,B):
    for i in range(len(A)):
        x1 = A[i]
        x2 = B[i] 
        x = np.array([1,x1,x2])
        w1=np.array([-1.5,1,1]) 
        if np.dot(x,w1) < 0.5:  
            print("A= " + str(A[i]) + " and " + "B= " + str(B[i]) + " = False")
        else:
            print("A= " + str(A[i]) + " and " + "B= " + str(B[i]) + " = True")


def boolean_Aand_notB (A,B):
    for i in range(len(A)):
        x1 = A[i]
        x2 = B[i]
        x = np.array([1,x1,x2])

        w1 = np.array([0,1,0])
        w2 = np.array([0,0,-1])

        x1 = np.dot(x,w1)
        x2 = np.dot(x,w2)
        x = np.array([1,x1,x2])

        w3 = np.array([-1.5,1,1]) 

        if np.dot(x,w3) < 0.5:
            print("A= " + str(A[i]) + " and " + "nicht B= " + str(B[i]) + " = False")
        else:
            print("A= " + str(A[i]) + " and " + "nicht B= " + str(B[i]) + " = True")

def boolean_AxorB (A,B):
    for i in range(len(A)):
        x1 = A[i]
        x2 = B[i]
        x = np.array([1,x1,x2])

        w1 = np.array([0,-1,1])
        w2 = np.array([0,-1,1])

        x1 = np.dot(x,w1)
        x2 = np.dot(x,w2)
        x = np.array([0,x1,x2])

        w3 = np.array([1,1,1]) 

        if np.dot(x,w3) == 0:
            print("A= " + str(A[i]) + " xor " + "B= " + str(B[i]) + " = False")
        else:
            print("A= " + str(A[i]) + " xor " + "B= " + str(B[i]) + " = True")
```
